```python
# Importing libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
store = pd.read_csv('SampleSuperstore.csv')
store.head()
```

```python
store.iloc[10]
```

```python
store.describe()
```

```python
store.hist(figsize=(12,17),bins=30);
```

```python
store.mean()
```

```python
# Correlation between the data
sns.heatmap(store.corr(), annot=True);
```

```python
store.corr().plot.bar();
```

```python
sns.countplot(x="Category", data=store).set_title("Most sold product line");
```

```python
sns.countplot(x="Sub-Category", data=store).set_title("Most sold Sub-Category");
plt.xticks(rotation=270);
```

```python
# Let's plot a relationship between sales and profit
plt.scatter(x=store['Sales'], y=store['Profit'])
plt.xlabel("Total Sales")
plt.ylabel("Profit")
plt.show()
```

```python
sns.set_style('ticks')
sns.countplot(x='Region',data=store,palette='hls');
```

```python
sns.countplot(x='Region',hue='Category',data=store,palette='dark')
```
