```python
# Importing the libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
from sklearn.cluster import KMeans
```

```python
data=pd.read_csv('iris.csv')
data.head()
```

```python
# Scatter plot eith sepallength and sepalwidth
plt.scatter(data['SepalLengthCm'],data['SepalWidthCm'])

plt.xlabel('Length of sepal')
plt.ylabel('Width of sepal')
plt.show()
```

```python
# Drop 'Id'---->> unecessary
data.drop(['Id'],inplace=True,axis=1)
```

```python
# Mapping the 'Species' with num
data['Species'] = data['Species'].map({'Iris-setosa':0, 'Iris-versicolor':1 , 'Iris-virginica':2})
```

```python
x=data.iloc[:,:-1]
y=data.iloc[:,-1]
```

```python
data.head()
```

```python
# Creating a variable which will contain the data for the clustering
x1= data.copy()
kmeans = KMeans(3)
kmeans.fit(x)
```

```python
# Create a copy of data, so we can see the clusters next to the original data
cluster = data.copy()
# predict the cluster for each observation
cluster['cluster_pred']=kmeans.fit_predict(x)
cluster
```

```python
plt.scatter(cluster['SepalLengthCm'], cluster['SepalWidthCm'], c= cluster ['cluster_pred'], cmap = 'rainbow')
plt.show()
```

```python
# Data preprocessing
from sklearn import preprocessing

# Scaling the data
x_scaled = preprocessing.scale(x)
```

```python
x_scaled
```

```python
# Creating a k-means object with 2 clusters
kmeans_scaled = KMeans(2)
kmeans_scaled.fit(x_scaled)
```

```python
# Creating a copy of data, so we can see the clusters next to the original data
cluster_scaled = data.copy()
cluster_scaled['cluster_pred']=kmeans_scaled.fit_predict(x_scaled)
```

```python
# Plotting the graph
plt.scatter(cluster_scaled['SepalLengthCm'], cluster_scaled['SepalWidthCm'], c= cluster_scaled ['cluster_pred'], cmap = 'rainbow')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.show()
```

```python
# Elbow method ---->>> wcss
wcss = []
# 'cl_num' --->>> keeps track the highest number of clusters we want to use the WCSS method for. We have it set at 10 right now, but it is completely arbitrary.
cl_num = 10
for i in range (1,cl_num):
    kmeans= KMeans(i)
    kmeans.fit(x_scaled)
    wcss_iter = kmeans.inertia_
    wcss.append(wcss_iter)
wcss
```

```python
number_cluster = range(1,cl_num)
plt.plot(number_cluster, wcss, marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Within-cluster Sum of Squares')
```

```python
# For 3 clusters
kmeans_3 = KMeans(3)
kmeans_3.fit(x_scaled)
```

```python
# Copy the data
cluster_3 = x.copy()
cluster_3['cluster_pred']=kmeans_3.fit_predict(x_scaled)
```
